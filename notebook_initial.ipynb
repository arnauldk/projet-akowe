```python
indicators =[
    "NY.GDP.PCAP.CD",  # GDP per capita (current US$) # PIB par habitant (dollars US courants)
    "SE.TER.ENRL",  # Enrolment in tertiary education, all programmes, both sexes (number ) # Inscriptions dans l'enseignement supérieur, tous programmes, les deux sexes (nombre)
    "IT.NET.USER.P2",    # Internet users (per 100 people) # Utilisateurs d'Internet (pour 100 personnes)
    "SE.XPD.TOTL.GB.ZS",     # Government expenditure on education (% of government expenditure)  # Dépenses publiques pour l'éducation (% des dépenses gouvernementales)
    "SP.POP.1564.TO.ZS",     # Population ages 15-64 (% of total) # Population âgée de 15 à 64 ans (% du total)
    "SP.POP.0014.TO.ZS",  # Population ages 0-14 (% of total) # Population âgée de 0 à 14 ans (% du total)
]

# Filtrer Afrique subsaherienne
ssa_codes= countries[countries["region"] == "Sub-Saharan Africa"]["country_code"]

# Filtrer Indicateurs + pays
df_ssa= df[
    (df["indicator_code"].isin(indicators)) &
    (df["country_code"].isin(ssa_codes))
]
df_ssa
```

```python
year_cols = [col for col in df_ssa.columns if col.startswith('col_') and col[4:].isdigit()]

df = df_ssa.melt(
    id_vars=['country_name', 'country_code', 'indicator_name', 'indicator_code'],
    value_vars=year_cols,
    var_name='Year',
    value_name='value'
)
df["Year"] = df["Year"].str.replace('col_', '').astype(int)
df["value"] = pd.to_numeric(df["value"], errors='coerce')
df.head()
```

```python
df = df[df["Year"].between(2010, 2024)]
df.head()
```

```python
df["value"] = pd.to_numeric(df["value"])
df = df.pivot_table(
    index=['country_name', "country_code", "Year" ],
    columns='indicator_name',
    values='value'
    ).reset_index()
df
```
